```python
%%sql
CREATE TABLE employee (
    Employee_ID INTEGER PRIMARY KEY AUTOINCREMENT,
    Name TEXT NOT NULL,
    Age INTEGER NOT NULL,
    Address TEXT,
    Salary REAL NOT NULL,
    Position TEXT,
    Department TEXT,
    Date_of_Joining DATE,
    Experience_Years INTEGER,
    Number_of_Leaves INTEGER,
    Performance_Rating REAL,
    Employee_Satisfaction REAL,
    Number_of_Promotions INTEGER,
    Number_of_Projects_Working INTEGER
);
```

Employee_ID:

    Description: A unique identifier for each employee.
    Data Type: INTEGER
    Constraints: PRIMARY KEY, AUTOINCREMENT
    Details: Automatically increments with each new record, ensuring uniqueness.

Name:

    Description: The full name of the employee.
    Data Type: TEXT
    Constraints: NOT NULL
    Details: Required field; should contain the employee's name.

Age:

    Description: The age of the employee.
    Data Type: INTEGER
    Constraints: NOT NULL
    Details: Required field; should contain the employee's age.

Address:

    Description: The residential address of the employee.
    Data Type: TEXT
    Constraints: None
    Details: Optional field; can contain the full address of the employee.

Salary:

    Description: The monthly or annual salary of the employee.
    Data Type: REAL
    Constraints: NOT NULL
    Details: Required field; should reflect the employee's salary.

Position:

    Description: The job title or role of the employee within the company.
    Data Type: TEXT
    Constraints: None
    Details: Optional field; should contain the employee's job title.

Department:

    Description: The department in which the employee works.
    Data Type: TEXT
    Constraints: None
    Details: Optional field; should reflect the department name.

Date_of_Joining:

    Description: The date when the employee joined the company.
    Data Type: DATE
    Constraints: None
    Details: Optional field; should be in a date format (YYYY-MM-DD).

Experience_Years:

    Description: The number of years of experience the employee has.
    Data Type: INTEGER
    Constraints: None
    Details: Optional field; should indicate the total years of experience.

Number_of_Leaves:

    Description: The total number of leaves taken by the employee.
    Data Type: INTEGER
    Constraints: None
    Details: Optional field; should reflect the number of leave days.

Performance_Rating:

    Description: A rating of the employee's performance.
    Data Type: REAL
    Constraints: None
    Details: Optional field; can be a numerical value representing performance.

Employee_Satisfaction:

    Description: The level of satisfaction the employee has with their job.
    Data Type: REAL
    Constraints: None
    Details: Optional field; can be a numerical value or percentage indicating satisfaction.

Number_of_Promotions:

    Description: The number of promotions the employee has received.
    Data Type: INTEGER
    Constraints: None
    Details: Optional field; should reflect the total number of promotions.

Number_of_Projects_Working:

    Description: The number of projects the employee is currently working on.
    Data Type: INTEGER
    Constraints: None
    Details: Optional field; should indicate the number of active projects.
